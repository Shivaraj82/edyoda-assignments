```python
import json
class Employee:
    def __init__(self, name, dob, height, city, state):
        self.name = name
        self.dob = dob
        self.height = height
        self.city = city
        self.state = state

# Function to read employee data from JSON file
def read_employee_data(file_name):
    try:
        with open(file_name, 'r') as file:
            employee_data = json.load(file)
        return employee_data
    except FileNotFoundError:
        print(f"File '{file_name}' not found.")
        return []

# Function to convert JSON data to Employee objects
def json_to_employee_objects(json_data):
    employees = []
    for emp_info in json_data:
        employee = Employee(
            emp_info['Name'],
            emp_info['DOB'],
            emp_info['Height'],
            emp_info['City'],
            emp_info['State']
        )
        employees.append(employee)
    return employees

# Main function
if __name__ == "__main__":
    # Create employee data JSON file
    employee_data = [
        {
            "Name": "virat Kohli",
            "DOB": "1990-09-27",
            "Height": 175,
            "City": "New delhi",
            "State": "Delhi"
        },
        {
            "Name": "rohit Sharma",
            "DOB": "1988-08-20",
            "Height": 173,
            "City": "Mumbai",
            "State": "Maharastra"
        },
        {
            "Name": "Shubman Gill",
            "DOB": "1997-03-10",
            "Height": 178,
            "City": "Chandigarh",
            "State": "Haryana"
        },
        {
            "Name": "KL Rahul",
            "DOB": "1992-11-25",
            "Height": 168,
            "City": "Banglore",
            "State": "Karnataka"
        },
        {
            "Name": "Ishan Kishan",
            "DOB": "1987-12-05",
            "Height": 173,
            "City": "Ranchi",
            "State": "Jharkand"
        }
    ]

    with open('employee.json', 'w') as file:
        json.dump(employee_data, file, indent=4)

    # Read and convert employee data to Employee objects
    employee_data = read_employee_data('employee.json')
    employees = json_to_employee_objects(employee_data)

    # Print the list of Employee objects
    for employee in employees:
        print(f"Name: {employee.name}")
        print(f"DOB: {employee.dob}")
        print(f"Height: {employee.height} cm")
        print(f"City: {employee.city}")
        print(f"State: {employee.state}")
        print()
```
